```python
import numpy as np
from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
```

```python
mp_key = "nbKtMcNjVctWQexVKEJTiT489Bo6oYrr"
mp_id = "mp-11869"
```

```python
#Récupération de la structure
with MPRester(mp_key) as m:
    structure = m.get_structure_by_material_id(mp_id)
```

## Diffractogramme

```python
xrd = XRDCalculator("CuKa1") #Pourquoi CuKa1? Car c'est celui qui correspond à la longueur d'onde λ=1.541Å selon Materials Project
pattern = xrd.get_pattern(structure)
```

Les pics sont caractérisés par un angle 2Θ entre les 2 rayons, une intensité, des indices hkl (plans correspondants) et la distance entre les plans. Dans la suite de ce notebook, on peut constater la notion de multiplicité des indices hkl. Cela est dû au fait que, pour un réseau cubique comme le notre (cf lattice.ipykernel), la distance entre les plans d_hkl dépend des indices au carré. De ce fait, un plan (010) et un plan (0-10) auront la même distance. La mutliplicité est donc le nombre de plan (hkl) ayant la même distance. 

### Premier pic

```python
print("Premier pic du diffractogramme pour un rayonnement X incident CuKα de longueur d’onde λ=1.54060 Å:")
print("Angle 2Θ :", pattern.x[0])
print("Intensity :", pattern.y[0])
print("hkl :", pattern.hkls[0])
print("distance :", pattern.d_hkls[0])
```
